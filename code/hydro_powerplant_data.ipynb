# Powerplant-data: Hydro

```python
import pandas as pd
```

```python
df = pd.read_csv("../pypsa-eur/data/powerplants/archive/0.8.1/powerplants.csv")
print("keys:" + str(df.keys()))
print("Fueltypes:" + str(df['Fueltype'].unique()))
print("Technologies:" + str(df['Technology'].unique()))
```

```python
hydro_res_de = df[
    (df['Country'] == 'Germany') & 
    (df['Fueltype'].str.contains('Hydro', case=False, na=False)) & 
    (df['Technology'] == 'Reservoir')
]
display(hydro_res_de.head())
hydro_res_de.to_csv('../results/hydro_res_de.csv', index=False)
```
